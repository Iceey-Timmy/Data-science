Assignment 3 -sets

```python
my_hobbies = {'coding','porno','music','research','reading'}
friend_hobbies = {'dancing','coding','travelling','porno','swimming'}
common_hobbies = my_hobbies.intersection(friend_hobbies)
unique_hobbies = my_hobbies.symmetric_difference(friend_hobbies)
print("common:",common_hobbies)
print("unique:",unique_hobbies)
```

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
print("Union:",A.union(B))
print("Difference(A-B):",A.difference(B))
```
